```python
import os
from os.path import isfile, join
import PIL.Image as pilimg
import numpy as np

output = []

##path = directory path##
class ImgToArr:    
    def __init__(self,path,row,length):
        
        #parameter       
        self.path= path
        self.row = row       #array size
        self.length = length #data size
    
    #threshold : pixel size >=128 : 0, pixel size<128 :1
    def threshold1(self,t):
            
        num = 0
        target = [] #labeling
            
        while num <self.length:
            
            self.dummy = np.zeros([self.row,self.row])    
            n = num+1
            a = pilimg.open(self.path+str(n)+".jpg")
            array = np.array(a)
            array = array[1:,1:] 
            
            for i in range(self.row):
                for j in range(self.row):
                    
                    if array[i,j]>=128:
                        self.dummy[i,j] = 0

                    else:
                        self.dummy[i,j] = 1       
                            
            output.append(self.dummy)
            target.append([t])
            num = num+1
            
                
        data = np.stack(output,axis=0)
        target = np.stack(target,axis = 0)
        
        return np.savez("HG_noise_dataset"+str(t)+".npz",data = data, target = target)
    
    #consider permutation 
    def threshold2(self,t):
            
        num = 0
        target = []
            
        while num <self.length:
                
            self.dummy = np.zeros([self.row,self.row])    
            n = num+1
            a = pilimg.open(self.path+str(n)+".jpg")
            array = np.array(a)
            array = array[1:,1:]
            
               
            for i in range(16):
                for j in range(15):
                    
                    #make first coulmn components 1 
                    self.dummy[i,0]=1
            
                    if array[i,j]>=128:
                        self.dummy[i,j+1] = 0
            
                    else:
                        self.dummy[i,j+1]=1
                            
            output.append(self.dummy)
            target.append([t])
            num = num+1
               
                
        data = np.stack(output,axis=0)
        target = np.stack(target,axis = 0)
        
        return np.savez("HG_noise_datasetP"+str(t)+".npz",data = data, target = target)
        
p='C:/Users/khl36/Hackathon/imgdata/HG/n=5,m=5 no.'

img = ImgToArr(p,16,100)
result = img.threshold1(35)

                        
```

```python
import numpy as np

class Dataftn:
    
    ##load npz file##
    def __init__(self,name,index):
        
        self.sample = np.load(name)
        data = self.sample['data']
        self.index = index
        self.data = data[index]
    
    ##take n-th index component of 2D array##
    def idx(self,x):
        
        self.x = x
        d = self.data
        data1d = d.flatten()
        
        return int(data1d[self.x])
    
a =  Dataftn('Dataset1.npz',0)
a.idx(125) 

    
    


```
